```python
from happytransformer import HappyGeneration
```

```python
happy_gen = HappyGeneration("GPT-NEO", "EleutherAI/gpt-neo-125M")
```

```python
from happytransformer import GENSettings
```

```python
args = GENSettings(no_repeat_ngram_size=2)
```

```python
top_k_sampling_settings = GENSettings(do_sample=True, early_stopping=False, top_k=50, temperature=0.1)
```

```python
result = happy_gen.generate_text("Changing mean twitts into nicer ones is to ", args=top_k_sampling_settings)
```

```python
print(result.text)
```

```python
import pandas as pd
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration,Adafactor
```

```python
if torch.cuda.is_available():
    dev = torch.device("cuda:0")
    print("Running on the GPU")
else:
    dev = torch.device("cpu")
    print("Running on the CPU")
```

```python
tokenizer = T5Tokenizer.from_pretrained("t5-base")
model = T5ForConditionalGeneration.from_pretrained("t5-base",
                                             return_dict=True)
#moving the model to GPU
model.to(dev)
```
